## 1. í´ë˜ìŠ¤ (Class)

## 2. ìº¡ìŠí™” (Encapsulation)

ìº¡ìŠí™”ëŠ” ë°ì´í„°ì™€ ë©”ì„œë“œë¥¼ í•˜ë‚˜ë¡œ ë¬¶ê³ , ì™¸ë¶€ì—ì„œ ì§ì ‘ ì ‘ê·¼ì„ ì œí•œí•˜ëŠ” ê°œë…ì…ë‹ˆë‹¤. Pythonì—ì„œëŠ” ì–¸ë”ìŠ¤ì½”ì–´(_)ë¥¼ ì‚¬ìš©í•´ ì ‘ê·¼ ì œì–´ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.

```python
class BankAccount:
    """ì€í–‰ ê³„ì¢Œ í´ë˜ìŠ¤ - ìº¡ìŠí™” ì˜ˆì œ"""
    
    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number  # ê³µê°œ ì†ì„±
        self._balance = initial_balance       # ë³´í˜¸ëœ ì†ì„± (protected)
        self.__pin = "1234"                   # ë¹„ê³µê°œ ì†ì„± (private)
    
    # Getter ë©”ì„œë“œ
    def get_balance(self):
        """ì”ì•¡ ì¡°íšŒ"""
        return self._balance
    
    # Setter ë©”ì„œë“œ
    def deposit(self, amount):
        """ì…ê¸ˆ"""
        if amount > 0:
            self._balance += amount
            print(f"{amount}ì›ì´ ì…ê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self._balance}ì›")
        else:
            print("ì…ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
    
    def withdraw(self, amount, pin):
        """ì¶œê¸ˆ (PIN í™•ì¸ í•„ìš”)"""
        if not self.__verify_pin(pin):
            print("PINì´ ì˜¬ë°”ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤.")
            return False
        
        if amount > self._balance:
            print("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤.")
            return False
        
        self._balance -= amount
        print(f"{amount}ì›ì´ ì¶œê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self._balance}ì›")
        return True
    
    def __verify_pin(self, pin):
        """PIN í™•ì¸ (private ë©”ì„œë“œ)"""
        return pin == self.__pin
    
    # Property ë°ì½”ë ˆì´í„° ì‚¬ìš©
    @property
    def balance(self):
        """ì”ì•¡ì„ ì†ì„±ì²˜ëŸ¼ ì ‘ê·¼í•  ìˆ˜ ìˆê²Œ í•¨"""
        return self._balance
    
    @balance.setter
    def balance(self, value):
        """ì”ì•¡ ì„¤ì • (ìœ íš¨ì„± ê²€ì‚¬ í¬í•¨)"""
        if value >= 0:
            self._balance = value
        else:
            print("ì”ì•¡ì€ ìŒìˆ˜ê°€ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
# ìº¡ìŠí™” ì‚¬ìš© ì˜ˆì œ
account = BankAccount("123-456-789", 10000)

# ê³µê°œ ë©”ì„œë“œë¥¼ í†µí•œ ì ‘ê·¼
print(f"ê³„ì¢Œë²ˆí˜¸: {account.account_number}")
print(f"ì”ì•¡: {account.get_balance()}ì›")

account.deposit(5000)
account.withdraw(3000, "1234")

# Propertyë¥¼ í†µí•œ ì ‘ê·¼
print(f"Propertyë¡œ ì ‘ê·¼í•œ ì”ì•¡: {account.balance}ì›")
account.balance = 20000
print(f"ì„¤ì • í›„ ì”ì•¡: {account.balance}ì›")

# ì˜ëª»ëœ ì ‘ê·¼ ì‹œë„
account.balance = -1000  # ìœ íš¨ì„± ê²€ì‚¬ì— ì˜í•´ ê±°ë¶€ë¨

# ë¹„ê³µê°œ ì†ì„±ì— ì§ì ‘ ì ‘ê·¼ ì‹œë„ (ì—ëŸ¬ ë°œìƒ)
try:
    print(account.__pin)
except AttributeError as e:
    print(f"ë¹„ê³µê°œ ì†ì„± ì ‘ê·¼ ì˜¤ë¥˜: {e}")
```

```python
# ë¶€ëª¨ í´ë˜ìŠ¤ (ê¸°ë³¸ í´ë˜ìŠ¤)
class Animal:
    """ë™ë¬¼ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    def __init__(self, name, species):
        self.name = name
        self.species = species
        self.is_alive = True
    
    def eat(self):
        """ë¨¹ê¸°"""
        print(f"{self.name}ì´(ê°€) ìŒì‹ì„ ë¨¹ê³  ìˆìŠµë‹ˆë‹¤.")
    
    def sleep(self):
        """ì ìê¸°"""
        print(f"{self.name}ì´(ê°€) ì ì„ ìê³  ìˆìŠµë‹ˆë‹¤.")
    
    def make_sound(self):
        """ì†Œë¦¬ë‚´ê¸° - ìì‹ í´ë˜ìŠ¤ì—ì„œ ì˜¤ë²„ë¼ì´ë“œë  ë©”ì„œë“œ"""
        print(f"{self.name}ì´(ê°€) ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.")
```

```python
# ìì‹ í´ë˜ìŠ¤ 1
class Dog(Animal):
    """ê°œ í´ë˜ìŠ¤ - Animalì„ ìƒì†"""
    
    def __init__(self, name, breed):
        super().__init__(name, "ê°œ")  # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ìƒì„±ì í˜¸ì¶œ
        self.breed = breed
    
    def make_sound(self):
        """ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©"""
        print(f"{self.name}ì´(ê°€) ë©ë© ì§–ìŠµë‹ˆë‹¤!")
    
    def fetch(self):
        """ê°œë§Œì˜ ê³ ìœ í•œ ë©”ì„œë“œ"""
        print(f"{self.name}ì´(ê°€) ê³µì„ ê°€ì ¸ì˜µë‹ˆë‹¤.")
    
    def wag_tail(self):
        """ê¼¬ë¦¬ í”ë“¤ê¸°"""
        print(f"{self.name}ì´(ê°€) ê¼¬ë¦¬ë¥¼ í”ë“­ë‹ˆë‹¤.")

# ìì‹ í´ë˜ìŠ¤ 2
class Cat(Animal):
    """ê³ ì–‘ì´ í´ë˜ìŠ¤ - Animalì„ ìƒì†"""
    
    def __init__(self, name, color):
        super().__init__(name, "ê³ ì–‘ì´")
        self.color = color
    
    def make_sound(self):
        """ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©"""
        print(f"{self.name}ì´(ê°€) ì•¼ì˜¹í•˜ê³  ì›ë‹ˆë‹¤.")
    
    def climb(self):
        """ê³ ì–‘ì´ë§Œì˜ ê³ ìœ í•œ ë©”ì„œë“œ"""
        print(f"{self.name}ì´(ê°€) ë‚˜ë¬´ë¥¼ íƒ€ê³  ì˜¬ë¼ê°‘ë‹ˆë‹¤.")
    
    def purr(self):
        """ê·¸ë¥´ë ê±°ë¦¬ê¸°"""
        print(f"{self.name}ì´(ê°€) ê·¸ë¥´ë ê±°ë¦½ë‹ˆë‹¤.")
```

```python
# ë‹¤ì¤‘ ìƒì† ì˜ˆì œ
class Bird(Animal):
    """ìƒˆ í´ë˜ìŠ¤"""
    
    def __init__(self, name, wingspan):
        super().__init__(name, "ìƒˆ")
        self.wingspan = wingspan
    
    def fly(self):
        """ë‚ ê¸°"""
        print(f"{self.name}ì´(ê°€) í•˜ëŠ˜ì„ ë‚ ì•„ê°‘ë‹ˆë‹¤.")
    
    def make_sound(self):
        print(f"{self.name}ì´(ê°€) ì§€ì €ê·‘ë‹ˆë‹¤.")

class Penguin(Bird):
    """í­ê·„ í´ë˜ìŠ¤ - Birdë¥¼ ìƒì†í•˜ì§€ë§Œ ë‚  ìˆ˜ ì—†ìŒ"""
    
    def __init__(self, name):
        super().__init__(name, 60)  # í­ê·„ì˜ í‰ê·  ë‚ ê°œ ê¸¸ì´
    
    def fly(self):
        """ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”© - í­ê·„ì€ ë‚  ìˆ˜ ì—†ìŒ"""
        print(f"{self.name}ì€(ëŠ”) ë‚  ìˆ˜ ì—†ì§€ë§Œ ìˆ˜ì˜ì„ ì˜í•©ë‹ˆë‹¤.")
    
    def swim(self):
        """ìˆ˜ì˜í•˜ê¸°"""
        print(f"{self.name}ì´(ê°€) ë¬¼ì†ì—ì„œ í—¤ì—„ì¹©ë‹ˆë‹¤.")
```

## 4. ì¶”ìƒí™” (Abstraction)

ì¶”ìƒí™”ëŠ” ë³µì¡í•œ êµ¬í˜„ ì„¸ë¶€ì‚¬í•­ì„ ìˆ¨ê¸°ê³  í•„ìš”í•œ ê¸°ëŠ¥ë§Œ ë…¸ì¶œí•˜ëŠ” ê°œë…ì…ë‹ˆë‹¤. Pythonì—ì„œëŠ” `abc` ëª¨ë“ˆì„ ì‚¬ìš©í•´ ì¶”ìƒ í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
from abc import ABC, abstractmethod
import math

# ì¶”ìƒ í´ë˜ìŠ¤ ì •ì˜
class Shape(ABC):
    """ë„í˜• ì¶”ìƒ í´ë˜ìŠ¤"""
    
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def area(self):
        """ë„“ì´ ê³„ì‚° - ë°˜ë“œì‹œ êµ¬í˜„í•´ì•¼ í•˜ëŠ” ì¶”ìƒ ë©”ì„œë“œ"""
        pass
    
    @abstractmethod
    def perimeter(self):
        """ë‘˜ë ˆ ê³„ì‚° - ë°˜ë“œì‹œ êµ¬í˜„í•´ì•¼ í•˜ëŠ” ì¶”ìƒ ë©”ì„œë“œ"""
        pass
    
    # êµ¬ì²´ì ì¸ ë©”ì„œë“œ (ëª¨ë“  ìì‹ í´ë˜ìŠ¤ì—ì„œ ê³µí†µìœ¼ë¡œ ì‚¬ìš©)
    def display_info(self):
        """ë„í˜• ì •ë³´ ì¶œë ¥"""
        print(f"ë„í˜•: {self.name}")
        print(f"ë„“ì´: {self.area():.2f}")
        print(f"ë‘˜ë ˆ: {self.perimeter():.2f}")
```

```python
# êµ¬ì²´ì ì¸ í´ë˜ìŠ¤ 1: ì›
class Circle(Shape):
    """ì› í´ë˜ìŠ¤"""
    
    def __init__(self, radius):
        super().__init__("ì›")
        self.radius = radius
    
    def area(self):
        """ì›ì˜ ë„“ì´ êµ¬í˜„"""
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        """ì›ì˜ ë‘˜ë ˆ êµ¬í˜„"""
        return 2 * math.pi * self.radius

# êµ¬ì²´ì ì¸ í´ë˜ìŠ¤ 2: ì§ì‚¬ê°í˜•
class Rectangle(Shape):
    """ì§ì‚¬ê°í˜• í´ë˜ìŠ¤"""
    
    def __init__(self, width, height):
        super().__init__("ì§ì‚¬ê°í˜•")
        self.width = width
        self.height = height
    
    def area(self):
        """ì§ì‚¬ê°í˜•ì˜ ë„“ì´ êµ¬í˜„"""
        return self.width * self.height
    
    def perimeter(self):
        """ì§ì‚¬ê°í˜•ì˜ ë‘˜ë ˆ êµ¬í˜„"""
        return 2 * (self.width + self.height)

# êµ¬ì²´ì ì¸ í´ë˜ìŠ¤ 3: ì‚¼ê°í˜•
class Triangle(Shape):
    """ì‚¼ê°í˜• í´ë˜ìŠ¤"""
    
    def __init__(self, side1, side2, side3):
        super().__init__("ì‚¼ê°í˜•")
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
    
    def area(self):
        """í—¤ë¡ ì˜ ê³µì‹ì„ ì‚¬ìš©í•œ ì‚¼ê°í˜• ë„“ì´ ê³„ì‚°"""
        s = self.perimeter() / 2  # ë°˜ë‘˜ë ˆ
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))
    
    def perimeter(self):
        """ì‚¼ê°í˜•ì˜ ë‘˜ë ˆ êµ¬í˜„"""
        return self.side1 + self.side2 + self.side3
```

```python
# ì¶”ìƒí™” ì‚¬ìš© ì˜ˆì œ
print("=== ì¶”ìƒí™” ì˜ˆì œ ===")

# ë‹¤ì–‘í•œ ë„í˜• ê°ì²´ ìƒì„±
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5)
]

# ì¶”ìƒí™”ë¥¼ í†µí•œ ì¼ê´€ëœ ì¸í„°í˜ì´ìŠ¤ ì‚¬ìš©
for shape in shapes:
    print()
    shape.display_info()
    print("-" * 30)

# ì¶”ìƒ í´ë˜ìŠ¤ëŠ” ì§ì ‘ ì¸ìŠ¤í„´ìŠ¤í™”í•  ìˆ˜ ì—†ìŒ
try:
    abstract_shape = Shape("ì¶”ìƒë„í˜•")
except TypeError as e:
    print(f"\nì¶”ìƒ í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤í™” ì˜¤ë¥˜: {e}")

# ì¶”ìƒ ë©”ì„œë“œë¥¼ êµ¬í˜„í•˜ì§€ ì•Šì€ í´ë˜ìŠ¤
class IncompleteShape(Shape):
    """ë¶ˆì™„ì „í•œ ë„í˜• í´ë˜ìŠ¤ - ì¶”ìƒ ë©”ì„œë“œë¥¼ êµ¬í˜„í•˜ì§€ ì•ŠìŒ"""
    
    def __init__(self):
        super().__init__("ë¶ˆì™„ì „í•œ ë„í˜•")
    
    # area()ì™€ perimeter() ë©”ì„œë“œë¥¼ êµ¬í˜„í•˜ì§€ ì•ŠìŒ

try:
    incomplete = IncompleteShape()
except TypeError as e:
    print(f"ë¶ˆì™„ì „í•œ êµ¬í˜„ ì˜¤ë¥˜: {e}")
```

## 5. ë‹¤í˜•ì„± (Polymorphism)

```python
# ë‹¤í˜•ì„±ì„ ìœ„í•œ ê¸°ë³¸ í´ë˜ìŠ¤
class Vehicle:
    """êµí†µìˆ˜ë‹¨ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def start(self):
        """ì‹œë™ ê±¸ê¸°"""
        print(f"{self.brand} {self.model}ì˜ ì‹œë™ì„ ê²ë‹ˆë‹¤.")
    
    def stop(self):
        """ì •ì§€í•˜ê¸°"""
        print(f"{self.brand} {self.model}ì´(ê°€) ì •ì§€í•©ë‹ˆë‹¤.")
    
    def move(self):
        """ì´ë™í•˜ê¸° - ìì‹ í´ë˜ìŠ¤ì—ì„œ ì˜¤ë²„ë¼ì´ë“œ"""
        print(f"{self.brand} {self.model}ì´(ê°€) ì´ë™í•©ë‹ˆë‹¤.")
```

```python
# ë‹¤ì–‘í•œ êµí†µìˆ˜ë‹¨ í´ë˜ìŠ¤ë“¤
class Car(Vehicle):
    """ìë™ì°¨ í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, fuel_type):
        super().__init__(brand, model)
        self.fuel_type = fuel_type
    
    def move(self):
        """ìë™ì°¨ì˜ ì´ë™ ë°©ì‹"""
        print(f"{self.brand} {self.model}ì´(ê°€) ë„ë¡œë¥¼ ë‹¬ë¦½ë‹ˆë‹¤. (ì—°ë£Œ: {self.fuel_type})")
    
    def honk(self):
        """ê²½ì  ìš¸ë¦¬ê¸°"""
        print(f"{self.brand} {self.model}ì´(ê°€) ë¹µë¹µ ê²½ì ì„ ìš¸ë¦½ë‹ˆë‹¤!")

class Airplane(Vehicle):
    """ë¹„í–‰ê¸° í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, max_altitude):
        super().__init__(brand, model)
        self.max_altitude = max_altitude
    
    def move(self):
        """ë¹„í–‰ê¸°ì˜ ì´ë™ ë°©ì‹"""
        print(f"{self.brand} {self.model}ì´(ê°€) í•˜ëŠ˜ì„ ë‚ ì•„ê°‘ë‹ˆë‹¤. (ìµœëŒ€ê³ ë„: {self.max_altitude}m)")
    
    def takeoff(self):
        """ì´ë¥™í•˜ê¸°"""
        print(f"{self.brand} {self.model}ì´(ê°€) ì´ë¥™í•©ë‹ˆë‹¤!")

class Ship(Vehicle):
    """ë°° í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, capacity):
        super().__init__(brand, model)
        self.capacity = capacity
    
    def move(self):
        """ë°°ì˜ ì´ë™ ë°©ì‹"""
        print(f"{self.brand} {self.model}ì´(ê°€) ë°”ë‹¤ë¥¼ í•­í•´í•©ë‹ˆë‹¤. (ì ì¬ëŸ‰: {self.capacity}í†¤)")
    
    def anchor(self):
        """ë‹» ë‚´ë¦¬ê¸°"""
        print(f"{self.brand} {self.model}ì´(ê°€) ë‹»ì„ ë‚´ë¦½ë‹ˆë‹¤.")
```

```python
# ë‹¤í˜•ì„±ì„ í™œìš©í•œ í•¨ìˆ˜
def operate_vehicle(vehicle):
    """êµí†µìˆ˜ë‹¨ì„ ì¡°ì‘í•˜ëŠ” í•¨ìˆ˜ - ë‹¤í˜•ì„± í™œìš©"""
    print(f"\n=== {vehicle.__class__.__name__} ì¡°ì‘ ===")
    vehicle.start()
    vehicle.move()  # ê° í´ë˜ìŠ¤ì—ì„œ ì˜¤ë²„ë¼ì´ë“œëœ ë©”ì„œë“œ í˜¸ì¶œ
    vehicle.stop()

def race_vehicles(vehicles):
    """êµí†µìˆ˜ë‹¨ ê²½ì£¼ - ë‹¤í˜•ì„± í™œìš©"""
    print("\nğŸ êµí†µìˆ˜ë‹¨ ê²½ì£¼ ì‹œì‘! ğŸ")
    for i, vehicle in enumerate(vehicles, 1):
        print(f"{i}ë²ˆì§¸ ì°¸ê°€ì:")
        vehicle.move()

# ë‹¤í˜•ì„± ì‚¬ìš© ì˜ˆì œ
print("=== ë‹¤í˜•ì„± ì˜ˆì œ ===")

# ë‹¤ì–‘í•œ êµí†µìˆ˜ë‹¨ ê°ì²´ ìƒì„±
vehicles = [
    Car("í˜„ëŒ€", "ì†Œë‚˜íƒ€", "ê°€ì†”ë¦°"),
    Airplane("ë³´ì‰", "747", 12000),
    Ship("ì‚¼ì„±", "ì»¨í…Œì´ë„ˆì„ ", 20000)
]

# ë‹¤í˜•ì„±ì„ í†µí•œ ì¼ê´€ëœ ì¡°ì‘
for vehicle in vehicles:
    operate_vehicle(vehicle)

# ë¦¬ìŠ¤íŠ¸ë¥¼ í†µí•œ ë‹¤í˜•ì„± í™œìš©
race_vehicles(vehicles)
```

```python
# ë©”ì„œë“œ ì˜¤ë²„ë¡œë”©ê³¼ ìœ ì‚¬í•œ íš¨ê³¼ (Pythonì—ì„œëŠ” ê¸°ë³¸ ë§¤ê°œë³€ìˆ˜ ì‚¬ìš©)
class Calculator:
    """ê³„ì‚°ê¸° í´ë˜ìŠ¤ - ë‹¤í˜•ì„± ì˜ˆì œ"""
    
    def add(self, a, b=None, c=None):
        """ë§ì…ˆ - ë§¤ê°œë³€ìˆ˜ ê°œìˆ˜ì— ë”°ë¼ ë‹¤ë¥´ê²Œ ë™ì‘"""
        if c is not None:
            return a + b + c
        elif b is not None:
            return a + b
        else:
            return a
    
    def multiply(self, *args):
        """ê³±ì…ˆ - ê°€ë³€ ë§¤ê°œë³€ìˆ˜ ì‚¬ìš©"""
        result = 1
        for num in args:
            result *= num
        return result

# ì—°ì‚°ì ì˜¤ë²„ë¡œë”© ì˜ˆì œ
class Vector:
    """ë²¡í„° í´ë˜ìŠ¤ - ì—°ì‚°ì ì˜¤ë²„ë¡œë”©"""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """+ ì—°ì‚°ì ì˜¤ë²„ë¡œë”©"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __mul__(self, scalar):
        """* ì—°ì‚°ì ì˜¤ë²„ë¡œë”© (ìŠ¤ì¹¼ë¼ ê³±)"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __str__(self):
        """ë¬¸ìì—´ í‘œí˜„"""
        return f"Vector({self.x}, {self.y})"
    
    def __eq__(self, other):
        """== ì—°ì‚°ì ì˜¤ë²„ë¡œë”©"""
        return self.x == other.x and self.y == other.y
```

```python
print("\n=== ê³„ì‚°ê¸° ë‹¤í˜•ì„± ===")
calc = Calculator()
print(f"add(5): {calc.add(5)}")
print(f"add(5, 3): {calc.add(5, 3)}")
print(f"add(5, 3, 2): {calc.add(5, 3, 2)}")
print(f"multiply(2, 3, 4): {calc.multiply(2, 3, 4)}")

print("\n=== ë²¡í„° ì—°ì‚°ì ì˜¤ë²„ë¡œë”© ===")
v1 = Vector(2, 3)
v2 = Vector(1, 4)
v3 = v1 + v2  # __add__ ë©”ì„œë“œ í˜¸ì¶œ
v4 = v1 * 3   # __mul__ ë©”ì„œë“œ í˜¸ì¶œ

print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v1 + v2: {v3}")
print(f"v1 * 3: {v4}")
print(f"v1 == v2: {v1 == v2}")
```

## 6. ì¢…í•© ì˜ˆì œ: ê²Œì„ ìºë¦­í„° ì‹œìŠ¤í…œ

ëª¨ë“  OOP ê°œë…ì„ í™œìš©í•œ ì¢…í•©ì ì¸ ì˜ˆì œì…ë‹ˆë‹¤.

```python
from abc import ABC, abstractmethod
import random

# ì¶”ìƒ ê¸°ë³¸ í´ë˜ìŠ¤
class GameCharacter(ABC):
    """ê²Œì„ ìºë¦­í„° ì¶”ìƒ í´ë˜ìŠ¤"""
    
    def __init__(self, name, health, attack_power):
        self._name = name           # ë³´í˜¸ëœ ì†ì„±
        self._health = health       # ë³´í˜¸ëœ ì†ì„±
        self._max_health = health   # ìµœëŒ€ ì²´ë ¥
        self._attack_power = attack_power
        self._level = 1
        self._experience = 0
    
    # Propertyë¥¼ í†µí•œ ìº¡ìŠí™”
    @property
    def name(self):
        return self._name
    
    @property
    def health(self):
        return self._health
    
    @property
    def level(self):
        return self._level
    
    @property
    def is_alive(self):
        return self._health > 0
    
    def take_damage(self, damage):
        """í”¼í•´ ë°›ê¸°"""
        self._health = max(0, self._health - damage)
        print(f"{self._name}ì´(ê°€) {damage}ì˜ í”¼í•´ë¥¼ ë°›ì•˜ìŠµë‹ˆë‹¤. (ì²´ë ¥: {self._health}/{self._max_health})")
        
        if not self.is_alive:
            print(f"{self._name}ì´(ê°€) ì“°ëŸ¬ì¡ŒìŠµë‹ˆë‹¤!")
    
    def heal(self, amount):
        """ì¹˜ë£Œí•˜ê¸°"""
        old_health = self._health
        self._health = min(self._max_health, self._health + amount)
        actual_heal = self._health - old_health
        print(f"{self._name}ì´(ê°€) {actual_heal}ë§Œí¼ íšŒë³µí–ˆìŠµë‹ˆë‹¤. (ì²´ë ¥: {self._health}/{self._max_health})")
    
    def gain_experience(self, exp):
        """ê²½í—˜ì¹˜ íšë“"""
        self._experience += exp
        print(f"{self._name}ì´(ê°€) {exp} ê²½í—˜ì¹˜ë¥¼ íšë“í–ˆìŠµë‹ˆë‹¤.")
        
        # ë ˆë²¨ì—… ì²´í¬
        if self._experience >= self._level * 100:
            self._level_up()
    
    def _level_up(self):
        """ë ˆë²¨ì—… (ë³´í˜¸ëœ ë©”ì„œë“œ)"""
        self._level += 1
        self._experience = 0
        self._max_health += 20
        self._health = self._max_health
        self._attack_power += 5
        print(f"ğŸ‰ {self._name}ì´(ê°€) ë ˆë²¨ {self._level}ë¡œ ë ˆë²¨ì—…í–ˆìŠµë‹ˆë‹¤!")
    
    @abstractmethod
    def attack(self, target):
        """ê³µê²© - ì¶”ìƒ ë©”ì„œë“œ"""
        pass
    
    @abstractmethod
    def special_ability(self):
        """íŠ¹ìˆ˜ ëŠ¥ë ¥ - ì¶”ìƒ ë©”ì„œë“œ"""
        pass
    
    def display_status(self):
        """ìƒíƒœ í‘œì‹œ"""
        print(f"--- {self._name} (ë ˆë²¨ {self._level}) ---")
        print(f"ì²´ë ¥: {self._health}/{self._max_health}")
        print(f"ê³µê²©ë ¥: {self._attack_power}")
        print(f"ê²½í—˜ì¹˜: {self._experience}/{self._level * 100}")
```

```python
# êµ¬ì²´ì ì¸ ìºë¦­í„° í´ë˜ìŠ¤ë“¤
class Warrior(GameCharacter):
    """ì „ì‚¬ í´ë˜ìŠ¤"""
    
    def __init__(self, name):
        super().__init__(name, 120, 25)
        self._armor = 10
    
    def attack(self, target):
        """ê¸°ë³¸ ê³µê²©"""
        if not self.is_alive:
            print(f"{self._name}ì€(ëŠ”) ì“°ëŸ¬ì ¸ì„œ ê³µê²©í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        damage = self._attack_power + random.randint(-5, 5)
        print(f"{self._name}ì´(ê°€) {target.name}ì„(ë¥¼) ê²€ìœ¼ë¡œ ê³µê²©í•©ë‹ˆë‹¤!")
        target.take_damage(damage)
        
        if not target.is_alive:
            self.gain_experience(50)
    
    def special_ability(self):
        """íŠ¹ìˆ˜ ëŠ¥ë ¥: ë°©ì–´ íƒœì„¸"""
        print(f"{self._name}ì´(ê°€) ë°©ì–´ íƒœì„¸ë¥¼ ì·¨í•©ë‹ˆë‹¤! ë°©ì–´ë ¥ì´ ì¦ê°€í•©ë‹ˆë‹¤.")
        self._armor += 5
        return f"{self._name}ì˜ ë°©ì–´ë ¥ì´ {self._armor}ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤."

class Mage(GameCharacter):
    """ë§ˆë²•ì‚¬ í´ë˜ìŠ¤"""
    
    def __init__(self, name):
        super().__init__(name, 80, 30)
        self._mana = 100
        self._max_mana = 100
    
    @property
    def mana(self):
        return self._mana
    
    def attack(self, target):
        """ê¸°ë³¸ ê³µê²©"""
        if not self.is_alive:
            print(f"{self._name}ì€(ëŠ”) ì“°ëŸ¬ì ¸ì„œ ê³µê²©í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        if self._mana < 10:
            print(f"{self._name}ì˜ ë§ˆë‚˜ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤!")
            return
        
        self._mana -= 10
        damage = self._attack_power + random.randint(-3, 8)
        print(f"{self._name}ì´(ê°€) {target.name}ì—ê²Œ íŒŒì´ì–´ë³¼ì„ ì‹œì „í•©ë‹ˆë‹¤!")
        target.take_damage(damage)
        
        if not target.is_alive:
            self.gain_experience(50)
    
    def special_ability(self):
        """íŠ¹ìˆ˜ ëŠ¥ë ¥: ë§ˆë‚˜ íšŒë³µ"""
        recovered = min(30, self._max_mana - self._mana)
        self._mana += recovered
        print(f"{self._name}ì´(ê°€) ëª…ìƒì„ í†µí•´ ë§ˆë‚˜ë¥¼ {recovered} íšŒë³µí–ˆìŠµë‹ˆë‹¤. (ë§ˆë‚˜: {self._mana}/{self._max_mana})")
        return f"ë§ˆë‚˜ {recovered} íšŒë³µ"
    
    def display_status(self):
        """ìƒíƒœ í‘œì‹œ (ì˜¤ë²„ë¼ì´ë“œ)"""
        super().display_status()
        print(f"ë§ˆë‚˜: {self._mana}/{self._max_mana}")

class Archer(GameCharacter):
    """ê¶ìˆ˜ í´ë˜ìŠ¤"""
    
    def __init__(self, name):
        super().__init__(name, 100, 20)
        self._arrows = 30
        self._accuracy = 0.8
    
    def attack(self, target):
        """ê¸°ë³¸ ê³µê²©"""
        if not self.is_alive:
            print(f"{self._name}ì€(ëŠ”) ì“°ëŸ¬ì ¸ì„œ ê³µê²©í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        if self._arrows <= 0:
            print(f"{self._name}ì˜ í™”ì‚´ì´ ë–¨ì–´ì¡ŒìŠµë‹ˆë‹¤!")
            return
        
        self._arrows -= 1
        
        if random.random() < self._accuracy:
            damage = self._attack_power + random.randint(0, 10)
            print(f"{self._name}ì´(ê°€) {target.name}ì„(ë¥¼) í™”ì‚´ë¡œ ëª…ì¤‘ì‹œì¼°ìŠµë‹ˆë‹¤!")
            target.take_damage(damage)
            
            if not target.is_alive:
                self.gain_experience(50)
        else:
            print(f"{self._name}ì˜ í™”ì‚´ì´ ë¹—ë‚˜ê°”ìŠµë‹ˆë‹¤!")
    
    def special_ability(self):
        """íŠ¹ìˆ˜ ëŠ¥ë ¥: ì •í™•ë„ í–¥ìƒ"""
        self._accuracy = min(0.95, self._accuracy + 0.1)
        print(f"{self._name}ì´(ê°€) ì§‘ì¤‘í•˜ì—¬ ì •í™•ë„ê°€ í–¥ìƒë˜ì—ˆìŠµë‹ˆë‹¤! (ì •í™•ë„: {self._accuracy:.0%})")
        return f"ì •í™•ë„ {self._accuracy:.0%}ë¡œ í–¥ìƒ"
```

```python
# ê²Œì„ ì‹œìŠ¤í…œ í´ë˜ìŠ¤
class GameSystem:
    """ê²Œì„ ì‹œìŠ¤í…œ ê´€ë¦¬ í´ë˜ìŠ¤"""
    
    def __init__(self):
        self.characters = []
    
    def add_character(self, character):
        """ìºë¦­í„° ì¶”ê°€"""
        self.characters.append(character)
        print(f"{character.name}ì´(ê°€) ê²Œì„ì— ì°¸ê°€í–ˆìŠµë‹ˆë‹¤!")
    
    def battle(self, char1, char2):
        """ì „íˆ¬ ì‹œìŠ¤í…œ"""
        print(f"\nâš”ï¸ {char1.name} vs {char2.name} ì „íˆ¬ ì‹œì‘! âš”ï¸")
        
        round_num = 1
        while char1.is_alive and char2.is_alive:
            print(f"\n--- ë¼ìš´ë“œ {round_num} ---")
            
            # í„´ ìˆœì„œ ê²°ì • (ëœë¤)
            if random.choice([True, False]):
                attacker, defender = char1, char2
            else:
                attacker, defender = char2, char1
            
            # ê³µê²©
            attacker.attack(defender)
            
            # ë°˜ê²© (ë°©ì–´ìê°€ ì‚´ì•„ìˆë‹¤ë©´)
            if defender.is_alive:
                defender.attack(attacker)
            
            round_num += 1
            
            # ë¬´í•œ ë£¨í”„ ë°©ì§€
            if round_num > 10:
                print("ì „íˆ¬ê°€ ë„ˆë¬´ ê¸¸ì–´ì ¸ì„œ ë¬´ìŠ¹ë¶€ë¡œ ëë‚¬ìŠµë‹ˆë‹¤.")
                break
        
        # ìŠ¹ì ê²°ì •
        if char1.is_alive and not char2.is_alive:
            print(f"ğŸ† {char1.name}ì´(ê°€) ìŠ¹ë¦¬í–ˆìŠµë‹ˆë‹¤!")
        elif char2.is_alive and not char1.is_alive:
            print(f"ğŸ† {char2.name}ì´(ê°€) ìŠ¹ë¦¬í–ˆìŠµë‹ˆë‹¤!")
    
    def display_all_characters(self):
        """ëª¨ë“  ìºë¦­í„° ìƒíƒœ í‘œì‹œ"""
        print("\n=== ëª¨ë“  ìºë¦­í„° ìƒíƒœ ===")
        for char in self.characters:
            char.display_status()
            print()
```

```python
# ê²Œì„ ì‹¤í–‰ ì˜ˆì œ
print("=== ê²Œì„ ìºë¦­í„° ì‹œìŠ¤í…œ ì¢…í•© ì˜ˆì œ ===")

# ê²Œì„ ì‹œìŠ¤í…œ ìƒì„±
game = GameSystem()

# ë‹¤ì–‘í•œ ìºë¦­í„° ìƒì„± (ë‹¤í˜•ì„±)
characters = [
    Warrior("ì•„ì„œ"),
    Mage("ê°„ë‹¬í”„"),
    Archer("ë ˆê³¨ë¼ìŠ¤")
]

# ìºë¦­í„°ë“¤ì„ ê²Œì„ì— ì¶”ê°€
for char in characters:
    game.add_character(char)

# ì´ˆê¸° ìƒíƒœ í‘œì‹œ
game.display_all_characters()

# íŠ¹ìˆ˜ ëŠ¥ë ¥ ì‚¬ìš© (ë‹¤í˜•ì„±)
print("\n=== íŠ¹ìˆ˜ ëŠ¥ë ¥ ì‚¬ìš© ===")
for char in characters:
    result = char.special_ability()
    print(f"ê²°ê³¼: {result}")

# ì „íˆ¬ ì‹œë®¬ë ˆì´ì…˜
warrior = characters[0]
mage = characters[1]
game.battle(warrior, mage)

# ìµœì¢… ìƒíƒœ í‘œì‹œ
game.display_all_characters()
```
